# Question 2: Maximum likelihood estimate (10 pts)

Consider the following univariate probability distribution 
$$p(x|a,b) = \frac{x^{a-1}}{\Gamma(a)b^a}exp(-\frac{x}{b}),$$ for $x \geq 0$ and $a,b > 0.  ~~ \Gamma(a)$ is a normalization term. 

You are given N i.i.d. data points drawn from this distribution. Derive the MLE estimate for $b$, in terms of $x_1, x_2, \ldots x_N$ and $a$.

## Answer



From the probability distribution, to get the MLE, we have to first take the sum of the log probabilities for all values of x. Then we take the partial derivative with respect to b, and then optimize to find the maximum. 

$ln(\Pi_{i=1}^N p(x_i|a,b))=\sum_{i=1}^N ln(p(x_i|a,b))$

Applying the log, we get

$\sum_{i=1}^N (a-1)ln(x_i) - ln(\Gamma(a)) - alnb - \frac{x_i}{b}$

Taking the partial derivative with respect to b:
$\frac{d}{db}=\sum_{i=1}^N \frac{x_i}{b^2}-\frac{a}{b}$

Setting this equal to 0, we find that
$-abN+\sum_{i=1}^N x_i = 0 \rightarrow \widehat{b}=\frac{\bar{x}}{a}$

The joint pdf of two random  variables $X_1$ and $X_2$ is  constant (i.e., the joint distribution is uniform) within the area shown below, and zero outside of it. Given this joint distribution, obtain  the following two univariate distributions from it. It is not necessary to write the equations, if the distributions can be plotted / described in full detail.

a. (6 pts) The marginal distribution of $X_1$, $p_{X_1}(x_1)$

b. (4 pts) The conditional distribution $p_{X_2|X_1}(x_2 \,|\, x_1 = 0.5)$

<img src="hw1p3_questionpic.png">

## Answer



### Part A
The marginal distribution of $X_1$ can be described as being proportional to the number of $X_2$ possibilities for each $X_1$.  Looking at this joint distribution, we can see that the marginal distribution is proportional to the height of the parallelogram at $X_1$.  In other words, the probability that $X_1$=1 is the largest probability, since the height of the parallelogram at that point is the maximum possible height.  As $X_1$ goes from 0 to 1, the probability increases, and as $X_1$ goes from 1 to 2, the probability decreases.  Thus, if we were to draw the probability distribution, it would look like an isosceles triangle, with width of 2, and height of 1.

### Part B

Since this is a bivariate uniform distribution, the conditional distribution of $X_2|X_1=0.5$ is represented by all possible $X_2$ points in and outside the parallelogram at $X_1=0.5$. We find the conditional distribution of $X_2|X_1=0.5$ to be 1 when $0\leq X_2\leq 1$, and 0 when $1\leq X_2 \leq 2$.

# Question 6: Illustrating the curse of dimensionality (10 pts)

For a hypersphere of radius $a$ in $d$ dimensions, the volume is related to the surface area of a unit hypersphere (S) as
    $$ V = \frac{S \times a^d}{d}.$$


1. (6 pts) Use this result to show that the fraction of the volume which lies at values of the radius between $a - \epsilon$ and $a$, where $0 < \epsilon < a$, is given by
$f = 1 - (1 - \epsilon/a)^d$. Hence, show that for any fixed $\epsilon$, no matter how small, this fraction tends to 1 as $d \rightarrow \infty$.
2. (4 pts) Evaluate the ratio $f$ numerically, with $\epsilon/a = 0.01$, for the cases $d$ = 2, 10, and 1000.


This problem shows that if data are randomly distributed within a very high-dimensional hypersphere, most of the mass is near the surface and there is very little of it near the "core". Thus, there are few "representative" points.

## Answer

1. Since $\epsilon < a$, we know that $\epsilon/a < 1$.  So, $1-\epsilon/a$ is also a fraction less than 1, which approaches 0 as d goes to infinity.  Hence, f goes to 1 as d goes to infinity.

2. $f=1 - (1-0.01)^2$ =0.0199; $1-(1-0.01)^{10}$ =0.0956; $1-(1-0.01)^{1000}$ =0.9999
